```python
#Loading libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
import ast

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df

#Clean data
df.job_posted_date = pd.to_datetime(df.job_posted_date)

df

```

```python
df.job_skills
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
df
```

```python
df.job_title_short.unique()
```

```python
# Draw chart showing top demanded skills for ALL data job
dfe=df.explode('job_skills')
dfe['job_skills'].value_counts().head().plot(kind = 'bar')
plt.show()
```

```python
data = df[df['job_title_short']=='Data Analyst'].explode('job_skills')
data['job_skills'].value_counts().head().plot(kind = 'bar')

```

```python
count = dfe.pivot_table(
    index = 'job_skills',
    columns = 'job_title_short',
    aggfunc='size'
)
job_requested = count['Senior Data Engineer'].reset_index(name = 'count')
job_requested.sort_values(by = 'count',ascending = False).head(3).plot(kind = 'barh',x='job_skills')
plt.gca().invert_yaxis()
plt.legend().set_visible(False)
```
