**MANAGE INDEX USING RESET, SET AND SORT INDEX**
* **df.reset_index():** renumber the index in the new dataframe
* **df.set_index():** bring the old version of index back
* **df.sort_values():** sort values in pivot tables (mostly used) in ascending order


```python
#Loading libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset

#Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df

#Clean data
df.job_posted_date = pd.to_datetime(df.job_posted_date)
```

```python
#Use df.reset_index() to renumber the index in the table
dfusa=df[df.job_country=="United States"]
dfusa.reset_index(inplace=True)
dfusa.head(3)
```

```python
#Using df.set_index() to bring the modified index to the original
dfusa.set_index('index').head(3)
```

```python
#Sort the values of yearly avarage salary in ascending order
dfusa_pivot=dfusa.pivot_table(
    values= 'salary_year_avg',
    index='job_title_short',
    aggfunc=['min','max','median']
)

dfusa_pivot.sort_values(by=('min','salary_year_avg'),ascending=False)
```
